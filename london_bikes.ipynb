```python
#read csv into pandas dataframe
bikes = pd.read_csv("london_merged.csv")
```

```python
#rename columns
new_cols_dict = {
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

bikes.rename(new_cols_dict, axis=1, inplace=True)
bikes.head()
```

```python
#changing humidity values to percentage
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
bikes.humidity_percent
```

```python
#Map season integers to strings
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

bikes.season = bikes.season.astype('str')
bikes.season = bikes.season.map(season_dict)
```

```python
#Map weather integers to strings
weather_dict = {
    '1.0':'clear',
    '2.0':'scattered clouds',
    '3.0':'broken clouds',
    '4.0':'cloudy',
    '7.0':'rain',
    '10.0':'rain with thunderstorm',
    '26.0':'snowfall'
}

bikes.weather = bikes.weather.astype('str')
bikes.weather = bikes.weather.map(weather_dict)
```

```python
#confirm mapping changes to season and weather
bikes.head()
```

```python
#export dataframe to an excel file for Tableau visualisation
bikes.to_excel('london_bikes_final.xlsx', sheet_name='Data')
```
